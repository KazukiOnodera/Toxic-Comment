```python
import pandas as pd
```

```python
cv = pd.read_csv('../output/test_fasttext6_cv_oof_0.988003_0.000558.csv')
train = pd.read_csv('../input/train.csv.zip')
```

```python
cv.head()
```

```python
cv = cv.add_prefix('cv_')
cv.columns = ['id'] + cv.columns[1:].tolist()
```

```python
df = pd.merge(train, cv, on='id', how='left')
```

```python
from sklearn.metrics import roc_auc_score
```

```python
roc_auc_score(df.toxic, df.cv_toxic)
```

```python
for c in ['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']:
    df['diff_'+c] = abs(df[c] - df['cv_'+c])
```

```python
df['diff_all'] = 0
for c in ['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']:
    df['diff_all'] += df['diff_'+c]
```
